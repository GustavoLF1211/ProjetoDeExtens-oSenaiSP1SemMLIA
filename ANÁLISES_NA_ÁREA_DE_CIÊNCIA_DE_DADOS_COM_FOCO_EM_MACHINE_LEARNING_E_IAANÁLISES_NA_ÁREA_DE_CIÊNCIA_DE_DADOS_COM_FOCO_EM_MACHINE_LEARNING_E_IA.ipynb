Na análise de dados, ele é crucial para estruturar séries temporais, permitindo extrair componentes específicos (ano, mês, dia da semana) e realizar análises baseadas em períodos

# **Tipos de Gráficos**

## **Gráfico de Barras**
Tipo de variável: Categóricas, tanto nominais quanto ordinais.

Permite Visualizar: Comparações entre categorias/Frequência de categorias.

Quando usar: Os gráficos de barras e colunas devem ser utilizados quando você tem dados classificados como nominais (dados categóricos que apresentam uma ordem natural) e ordinais (dados categóricos que não apresentam uma ordem natural).

### **Exemplo**

```python
import seaborn as sns

traducao = {
    'Entry-level': 'Júnior',
    'Mid-level': 'Pleno',
    'Senior': 'Sênior',
    'Executive': 'Executivo'
}

df_pt_br['nivel_experiencia_traduzido'] = df_pt_br['nivel_experiencia'].replace(traducao)

frequencia = df_pt_br['nivel_experiencia_traduzido'].value_counts()
```

```python
ordem_desejada = ['Júnior', 'Pleno', 'Sênior', 'Executivo']

plt.figure(figsize=(8,6))

sns.barplot(data=df_pt_br, order=ordem_desejada)

plt.bar(frequencia.index, frequencia.values)

plt.title('Frequência de Vagas por Nível de Experiência')
plt.xlabel('Nível de experiência')
plt.ylabel('Frequência')
plt.show()
```

**Explicação**

- .value_counts() - Conta a frequência de cada categoria na coluna.
- frequencia - Armazena 'int', pois é onde a frequência realizada pelo .value_counts() é armazenada.
- (frequencia.index, frequencia.values) - Correspondem a categorias e frequências (Categorias e altura do gráfico)
- plt.bar() - Esse comando cria o gráfico
- plt.title(), plt.xlabel(), plt.ylabel() - Coloca o nome do título, eixo x e eixo y.
- plt.show() - Mostra o gráfico na tela


**Ordenar**
- Para isso, armazene as categorias em uma lista, use o comando: sns.barplot(data=df_pt_br, order=ordem_desejada), onde você colocará o DataFrame e a lista na sequência.

## **Gráfico de Linhas**

Tipo de variável: Variáveis numéricas ao longo de um período de tempo

Permite visualizar: Evolução dos dados, tendências de mercado, variação em um período de tempo.

Quando usar: Ele conecta pontos de dados individuais com segmentos de reta, facilitando a identificação rápida de aumentos, diminuições, picos ou estabilização de métricas.

### **Exemplo**

```python
media_salario = df_pt_br.groupby('ano_trabalho')['salario_dolar'].mean()

plt.figure(figsize=(11, 6))
plt.plot(media_salario.index, media_salario.values, marker='o')

plt.title("Mediana Salarial Por Ano")
plt.xlabel("Ano")
plt.ylabel("Salario em USD")
plt.show()
```

**Explicação**
- .groupby('ano_trabalho') - Agrupa o ano de trabalho com o salário em dolár para o cálculo da mediana.
- mean() é para média, median() é mediana. Podemos utilizar mean em um caso onde os dados são muito bem distribuídos, porém a mediana é melhor no nosso caso, pois existem outliers e os dados são predominantes em uma faixa e poucos em outras.
- .index - Representa os valores presentes no eixo x.
- .values - Representa os valores no eixo y.
- É importante agregar dados por ano para simplificar os dados em um fluxo temporal, identificando melhor picos, quedas ou estabilidades de períodos.

### **Discussão**

- O salário mediano apresenta aumento ao longo dos anos.
- O gráfico é adequado para analisar médias, modas e medianas ao longo de um período de tempo determinado.
- Fatores externos que podem afetar a evolução: Popularidade da área de dados, possibilidade de home office durante a pandemia, procura de empregos relacionados a tecnologia, alta demanda por profissionais de dados nas empresas.

## **Gráfico de Pizza**

```python
media_salario_por_ano = df_pt_br.groupby(df_pt_br['ano_trabalho'].dt.year)['salario_dolar'].mean().sort_values()
percentage_increase = media_salario_por_ano.pct_change() * 100

jobs_ml_ia_por_ano = df_ml_ia['ano_trabalho'].dt.year.value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))


sns.barplot(
    x=media_salario_por_ano.index,
    y=media_salario_por_ano.values,
    palette='viridis',
    ax=axes[0]
)

# axes[0].set_title('Média Salarial Anual')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Salário (USD)')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Anotações %
for i, (year, value) in enumerate(media_salario_por_ano.items()):
    if not pd.isna(percentage_increase[year]):
        axes[0].text(
            i,
            value + (media_salario_por_ano.max() * 0.01),
            f'+{percentage_increase[year]:.1f}%',
            ha='center',
            fontsize=9
        )

axes[0].tick_params(axis='x', rotation=45)


sns.lineplot(
    x=jobs_ml_ia_por_ano.index,
    y=jobs_ml_ia_por_ano.values,
    marker='o',
    color='purple',
    ax=axes[1]
)

# axes[1].set_title('Crescimento de Cargos ML/IA')
axes[1].set_xlabel('Ano')
axes[1].set_ylabel('Número de Cargos')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

axes[1].set_xticks(jobs_ml_ia_por_ano.index)


plt.tight_layout()
plt.show()
```
